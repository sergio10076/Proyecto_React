```python
import math
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/sergio10076/pandas/main/Distancia10.csv'
url1 = 'https://raw.githubusercontent.com/sergio10076/pandas/main/Velocidades.csv'
url2 = 'https://raw.githubusercontent.com/sergio10076/pandas/main/nombres.csv'
df = pd.read_csv(url, sep=";",header=None)
df1 = pd.read_csv(url1, sep=";",header=None)
df2 = pd.read_csv(url2, sep=";",header=None)
m_distancias = df.to_numpy()
m_velocidad = df1.to_numpy()
nombres = df2.to_numpy()
N=int(math.sqrt(m_distancias.size))
matriz = np.empty((N,N))
for i in range(N):
  for j in range(N):
        if m_velocidad[i][j] == 0:
          matriz[i][j] = 0
        else:
          matriz[i][j]=m_distancias[i][j]/m_velocidad[i][j]   

maximizar = float('inf')
  
def copiarAlFinal(ruta_actual):
    ruta_final[:N + 1] = ruta_actual[:]
    ruta_final[N] = ruta_actual[0]
  
def pri_min(matriz, i):
    min = maximizar
    for k in range(N):
        if matriz[i][k] < min and i != k:
            min = matriz[i][k]
  
    return min
  
def se_min(matriz, i):
    primero, segundo = maximizar, maximizar
    for j in range(N):
        if i == j:
            continue
        if matriz[i][j] <= primero:
            segundo = primero
            primero = matriz[i][j]
  
        elif(matriz[i][j] <= segundo and 
             matriz[i][j] != primero):
            segundo = matriz[i][j]
  
    return segundo
  

def Recorrido_visitador(matriz, lim_inf, pe_ruta, 
              nivel, alm_solu, visitado):
    global final_res
      
    if nivel == N:

        if matriz[alm_solu[nivel - 1]][alm_solu[0]] != 0:
              
            
            pe_total = pe_ruta + matriz[alm_solu[nivel - 1]]\
                                        [alm_solu[0]]
            if pe_total < final_res:
                copiarAlFinal(alm_solu)
                final_res = pe_total
        return
  
  
    for i in range(N):
          

        if (matriz[alm_solu[nivel-1]][i] != 0 and
                            visitado[i] == False):
            temp = lim_inf
            pe_ruta += matriz[alm_solu[nivel - 1]][i]
  
           
            if nivel == 1:
                lim_inf -= ((pri_min(matriz, alm_solu[nivel - 1]) + 
                                pri_min(matriz, i)) / 2)
            else:
                lim_inf -= ((se_min(matriz, alm_solu[nivel - 1]) +
                                 pri_min(matriz, i)) / 2)
  
           
            if lim_inf + pe_ruta < final_res:
                alm_solu[nivel] = i
                visitado[i] = True
               
                Recorrido_visitador(matriz, lim_inf, pe_ruta, 
                       nivel + 1, alm_solu, visitado)

            pe_ruta -= matriz[alm_solu[nivel - 1]][i]
            lim_inf = temp
  
            visitado = [False] * len(visitado)
            for j in range(nivel):
                if alm_solu[j] != -1:
                    visitado[alm_solu[j]] = True

def Recorrido(matriz):
      

    lim_inf = 0
    alm_solu = [-1] * (N + 1)
    visitado = [False] * N
  
    for i in range(N):
        lim_inf += (pri_min(matriz, i) + 
                       se_min(matriz, i))

    lim_inf = math.ceil(lim_inf / 2)
  

    visitado[0] = True
    alm_solu[0] = 0
  

    Recorrido_visitador(matriz, lim_inf, 0, 1, alm_solu, visitado)
  

ruta_final = [None] * (N + 1)

visitado = [False] * N
  

final_res = maximizar
  
Recorrido(matriz)
```

```python
print("Tiempo de el recorrido mas eficiente :",round(final_res,2),"hrs")
print("Camino tomado : ")

for i in range(N+1):
  print(nombres[0][ruta_final[i]])

print("El promedio por entrevista no debe ser mayor a :", int(((8-final_res)/19)*60))
t_entrevistas = int(input("Tiempo en minutos para cada entrevista "))
print("El tiempo que se gasta en hacer las visistas y regresar es de: ", round(((t_entrevistas/60)*19)+final_res,2),"hrs")
```
